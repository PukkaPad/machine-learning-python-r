# Simple Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

%matplotlib inline
```

```python
dataset = pd.read_csv('../../data/Salary_Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:,1].values
```

### Training set and Test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1.0/3, random_state = 0)
```

```python
# it's not necessary to apply feature scaling for Simple Linear Regression, as the library will tak care of it
```

### Fitting Simple Linear Regression to the Training set

```python
regressor = LinearRegression().fit(X_train, y_train)
```

### Predicting the Test set results

```python
# Create a vector of predicted salaries
# I can compare how different the predicted salary is from the test set
```

```python
y_pred = regressor.predict(X_test)
```

### Visualising the Training set results

```python
# observation
plt.scatter(X_train, y_train, color = 'red', label = 'Observed (Train set)') 
# I will be the prediction on the training set 
# to compare real salary on the predicted salaries, based on the same observation (X_train)
plt.plot(X_train, regressor.predict(X_train), color = 'blue', label = 'Predicted')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
```

### Visualising the Test set results:

```python
plt.scatter(X_test, y_test, color = 'red', label = 'Observed (Test set)')
plt.plot(X_train, regressor.predict(X_train), color = 'blue', label = 'Predicted')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
```
